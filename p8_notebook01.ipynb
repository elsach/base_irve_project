### Pr√©vision des prochaines immatriculations de voitures √©lectriques √† 2 ans
L'historique reste de petite taille, id√©alement une granularit√© plus fine /Mois serait plus appropri√©e. Notons malgr√© tout qu'une pr√©vision √† court terme reste r√©alisable, essayons l'Algorithme Prophet (from Facebook).

```python
#Rappel du jeu de donn√©es utilis√© √† des fins de pr√©vision
immat_annee
```

Notons que l'√©chantillon pris pour r√©f√©rence est celui des immatriculations / ann√©e en valeurs non cumul√©es. Toutes les pr√©visions seront faites √† partir des valeurs non cumul√©es, ensuite sur Tableau les graphiques seront d√©taill√©s en cumul√© si n√©cessaire.

```python
#Pr√©paration des donn√©es et chargement de la librairie Prophet
from fbprophet import Prophet
plt.style.use('ggplot')
import warnings
warnings.filterwarnings("ignore")

forecast_prophet_ve = immat_annee.copy()
forecast_prophet_ve = forecast_prophet_ve[['Ann√©e', 'Electricit√©']]

forecast_prophet_ve.rename(columns={'Ann√©e': 'ds', 'Electricit√©': 'y'}, inplace=True)
forecast_prophet_ve
```

```python
#Sauvegarde 
forecast_prophet_ve.to_csv('p8_datatable/forecast_prophet_ve.csv')
```

```python
#Instanciation et entrainement du mod√®le
model = Prophet(weekly_seasonality=False, daily_seasonality=False,  yearly_seasonality=True, n_changepoints=7)
model.fit(forecast_prophet_ve)
```

```python
#Pr√©vision de l'√©volution des immatriculations des V√©hicules √©lectriques √† 2 ans 
future = model.make_future_dataframe(periods=2, freq='Y')
forecast = model.predict(future)
fig = model.plot(forecast)
fig.savefig('p8_img/forecast_prophet_ve.png')
```

```python
#Visualisation des donn√©es pr√©dictives
forecast_ve = model.predict(future)
forecast_ve[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
#Pr√©paration des donn√©es (observations + pr√©visions) pour Test statistique
metric_forecast_ve = forecast.set_index('ds')[['yhat', 'yhat_lower', 'yhat_upper']]\
                     .join(forecast_prophet_ve.set_index('ds')).reset_index()

metric_forecast_ve
```

```python
#Sauvegarde 
metric_forecast_ve.to_csv('p8_datatable/metric_forecast_ve.csv')
```

```python
#Import des fonctions statistiques from Scikit Learn
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
periods=2 #Pr√©vision ici sur 2 ans
```

```python
#Calcul du Coef. de d√©termination
r2_score(metric_forecast_ve['y'][:-periods], metric_forecast_ve['yhat'][:-periods])
```

Tr√®s bien! Mais voyons les autres coefficients pour avoir une id√©e plus pr√©cise.

```python
#Calcul du MSE
mean_squared_error(metric_forecast_ve['y'][:-periods], metric_forecast_ve['yhat'][:-periods])
```

```python
#Calcul du RMSE
math.sqrt(mean_squared_error(metric_forecast_ve['y'][:-periods], metric_forecast_ve['yhat'][:-periods]))
```

Une valeur proche de z√©ro confirme la robustesse du mod√®le, or ici la valeur est tr√®s importante. L'√©chantillon est de tr√®s petite taille, on peut en d√©duire qu'une pr√©diction ne sera pas fiable dans le temps, les craintes de d√©part semblent se confirmer, une granularit√© par mois aurait √©t√© pr√©f√©rable.

Maintenant ... voyons √† quoi ressemble l'erreur absolue moyenne (MAE).

```python
#Calcul du MAE
mean_absolute_error(metric_forecast_ve['y'][:-periods], metric_forecast_ve['yhat'][:-periods])
```

Les donn√©es √† disposition ne sont pas suffisantes pour pouvoir √©tablir une mod√©lisation fiable dans le temps. Notons que les pr√©visions sur une ou deux ann√©es semblent aller dans le bon sens, mais dans le contexte rencontr√© il n'est pas possible de pouvoir √©tablir d'autres pr√©visions √† plus long terme.

#### Analyse compl√©mentaire a posteriori
On tronque la s√©rie de l‚Äôann√©e 2019, qu‚Äôon cherche ensuite √† pr√©voir √† partir de l‚Äôhistorique 2010-2017. Cette analyse permet d'avoir une id√©e compl√©mentaire de la qualit√© pr√©dictive du mod√®le choisi.

```python
#Donn√©es tronqu√©e de l'ann√©e 2019
forecast_prophet_ve_trunc = forecast_prophet_ve[:-1]
```
